```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('zomato.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.drop(['url', 'address', 'phone', 'dish_liked', 'menu_item', 'reviews_list'], axis = 1)
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace = True)
df.shape
```

```python
df['rate'].unique()
```

```python
# remove   'NEW' , "-" , 'nan' , "/5" from rate column
```

```python
def handle_rate(value) :
    if value == 'NEW' or value == '-' :
        return np.nan
    else :
        value = str(value).split('/')
        value = value[0]
        return float(value)
        
df['rate'] = df['rate'].apply(handle_rate)        
df['rate'].head()
```

```python
df['rate'].isnull().sum()
```

```python
# filling null values in rate columns with mean

df['rate'].fillna(df['rate'].mean() , inplace = True)
df['rate'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
df.head()
```

```python
# outlier 
```

```python
df.rename(columns = {'approx_cost(for two people)' : "Cost2plate"  , 'listed_in(type)' : "Type" }, inplace = True )
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

```python
# location and  'listed_in(city)'  both are indicating location so drop one

df = df.drop('listed_in(city)',axis=1)
```

```python
df.head()
```

```python
df['Cost2plate'].unique()
```

```python
# removing ',' from cost2plate


def handle_comma(value) :
    value = str(value)
    if ',' in value :
        value = value.replace(",","")
        return float(value)
    else :
        return float(value)
    
df['Cost2plate'] =  df['Cost2plate'].apply(handle_comma)
```

```python
df['Cost2plate'].unique()
```

```python
df.head()
```

```python
df['rest_type'].value_counts(ascending = False)
```

```python
# grouping the rest_types (rare)   as others   (grouping less than 1000 restaurants as other rest_type)
```

```python
rest_type = df['rest_type'].value_counts(ascending = False)
rest_type
```

```python
rest_types_less_than_1000 = rest_type[rest_type < 1000]
rest_types_less_than_1000
```

```python
# making the rest type with less than 1000 frequency as others
```

```python
def handle_rest_type(value) :
    if value in rest_types_less_than_1000 :
        return 'others'
    else :
        return value
    
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
```

```python
df['rest_type'].value_counts()
```

```python
df.head()
```

```python
df['location'].value_counts()
```

```python
# do same make the low frequency location as others 
```

```python
location = df['location'].value_counts(ascending = False)
location
```

```python
location_less_than_300 = location[location < 300]
```

```python
def handle_location(value) :
    if value in location_less_than_300 :
        return 'others'
    else :
        return value
    
df['location'] = df['location'].apply(handle_location)   
df['location'].value_counts()
```

```python
df.head()
```

```python
# cleaning cusine 

df['cuisines'].value_counts()
```

```python
df['cuisines'].nunique()
```

```python
# grouping as other 

cuisines = df['cuisines'].value_counts()

cuisines
```

```python
cuisines_lessthan_300 = cuisines[cuisines < 300]
```

```python
def handle_cusine(value) :
    if value in cuisines_lessthan_300 :
        return 'others'
    else :
        return value

df['cuisines'] = df['cuisines'].apply(handle_cusine)       

df['cuisines'].value_counts()
```
