```python
#importing the libraries needed 
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from time import sleep
from random import randint
```

```python
#Declaring the headers 
headers = {"Accept-Language": "en-US,en;q=0.5"}
```

```python
movie_name = []
year = []
time=[]
rating=[]
meta_score =[]
votes = []
gross = []

```

https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating
    
    

https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&count=100&start=101&ref_=adv_nxt

https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&count=100&start=201&ref_=adv_nxt
    
https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&count=100&start=301&ref_=adv_nxt

```python
pages = np.arange(1,1000,100)

for i in pages :
    i = requests.get("https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&count=100&start="+str(i)+"&ref_=adv_nxt")
    soup = BeautifulSoup(i.text, 'html.parser')
    
    movie_data = soup.findAll('div',  attrs = {"class"  : "lister-item mode-advanced"})
    sleep(randint(2,8))  # for next  request  time gap  2-8 sec
    
    for j in movie_data :
        name = j.h3.a.text
        movie_name.append(name)
        
        run_time = j.p.find('span',class_ = 'runtime').text
        time.append(run_time)
        
        year_release = j.h3.find("span" , class_ = "lister-item-year text-muted unbold").text.replace('(', '').replace(')', '')
        year.append(year_release)
        
        r = j.find('div', class_ = 'inline-block ratings-imdb-rating').text.replace('\n',"")
        rating.append(r)
        
        meta  = j.find('span', class_ = 'metascore').text.replace(' ','')    if j.find('span', class_ = 'metascore') else 'null'
        meta_score.append(meta)
    
    # extract votes and gross 
        values = j.find_all('span', attrs = {'name' : 'nv'})
    
        vote  = values[0].text
        votes.append(vote)
    
        g = values[1].text if len(values) > 1 else  "null"     # (if dont use if condition then throws out  of range error)
        gross.append(g) 
    
    
    
    
       
   
```
